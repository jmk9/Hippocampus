```python
import os, glob
import numpy as np
import nibabel as nib
from skimage.transform import resize
```

```python
hdr_X1 = list()
X_root = '/content/Train'
y_root = '/content/Train/Labels'
root_hdr_X = glob.glob(os.path.join(X_root, '*.hdr'))
root_hdr_X.sort()
# img_X = list()
hdr_y1 = list()
root_hdr_y = glob.glob(os.path.join(y_root, '*.hdr'))
root_hdr_y.sort()
# img_y = list()

for idx, i in enumerate(root_hdr_X):
  # print(i)
  hdr1 = nib.load(i).get_fdata()
  hdr1 = resize(hdr1, (256, 256, 256))
  hdr1 = hdr1 / hdr1.max()
  hdr1 = hdr1 * 256.
  hdr_X1.append(hdr1)

for idx, i in enumerate(root_hdr_y):
  # print(i)
  hdr1 = nib.load(i).get_fdata()
  hdr1 = resize(hdr1, (256, 256, 256))
  hdr1 = hdr1 / hdr1.max()
  hdr1 = hdr1 * 256.
  hdr_y1.append(hdr1)
```

```python
resize_hdr_X = list()
resize_hdr_y = list()

for i in hdr_X1:
  re_img = i.reshape(i.shape[0],i.shape[1],i.shape[2])
  # print(re_img.shape)
  resize_hdr_X.append(re_img)

for i in hdr_y1:
  re_img = i.reshape(i.shape[0],i.shape[1],i.shape[2])
  # print(re_img.shape)
  resize_hdr_y.append(re_img)
```

```python
print(resize_hdr_X[-1].shape)
print(resize_hdr_y[-1].shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(resize_hdr_X, resize_hdr_y, test_size=0.2)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

img_size = resize_hdr_X[0].shape
num_classes = 2

inputs = keras.Input(shape=img_size + (1,))
x = layers.Rescaling(1./255)(inputs)
x = layers.Conv3D(64, 3, strides=2, activation="relu", padding="same")(x)
x = layers.Conv3D(64, 3, activation="relu", padding="same")(x)
x = layers.Conv3D(128, 3, strides=2, activation="relu", padding="same")(x)
x = layers.Conv3D(128, 3, activation="relu", padding="same")(x)
x = layers.Conv3D(256, 3, strides=2, padding="same", activation="relu")(x)
x = layers.Conv3D(256, 3, activation="relu", padding="same")(x)

x = layers.Conv3DTranspose(256, 3, activation="relu", padding="same")(x)
x = layers.Conv3DTranspose(256, 3, activation="relu", padding="same", strides=2)(x)
x = layers.Conv3DTranspose(128, 3, activation="relu", padding="same")(x)
x = layers.Conv3DTranspose(128, 3, activation="relu", padding="same", strides=2)(x)
x = layers.Conv3DTranspose(64, 3, activation="relu", padding="same")(x)
x = layers.Conv3DTranspose(64, 3, activation="relu", padding="same", strides=2)(x)

outputs = layers.Conv3D(num_classes, 3, activation="sigmoid", padding="same")(x)

model = keras.Model(inputs, outputs)

model.summary()
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc', 'mae'])
```

```python
from tensorflow.keras.callbacks import *

callbacks = [EarlyStopping(patience=5,
                           monitor='val_loss'),
             ModelCheckpoint('hippo_seg.keras',
                             save_best_only=True)]
```

```python
histroy = model.fit(X_train, y_train,
                    epochs=50,
                    callbacks=callbacks,
                    batch_size=2,
                    validation_data=(X_val, y_val))
```
